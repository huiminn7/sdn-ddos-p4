Load dataset

```python
import pandas as pd

# your SDN dataset
df_sdn = pd.read_csv("/home/huimin/testing/p4/ddos/dataset/traffic_log.csv")

# public dataset (cleaned + balanced)
df_pub = pd.read_csv("/home/huimin/testing/p4/ddos/dataset/public_balanced_clean.csv")
```

```python
features = [
    "packet_rate", "byte_rate", "avg_packet_size",
    "tcp_count", "udp_count", "icmp_count", "syn_count",
    "tcp_ratio", "udp_ratio", "icmp_ratio",
    "attack_type"
]

sdn_clean = df_sdn[features].dropna()
pub_clean = df_pub[features].dropna()

X_sdn = sdn_clean[features[:-1]]
y_sdn = sdn_clean["attack_type"]

X_pub = pub_clean[features[:-1]]
y_pub = pub_clean["attack_type"]
```

Scale the data

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

scaler = StandardScaler()

X_all = pd.concat([X_sdn, X_pub], ignore_index=True)
X_scaled = scaler.fit_transform(X_all)
X_scaled = pd.DataFrame(X_scaled, columns=X_sdn.columns)

X_sdn_scaled = X_scaled.iloc[:len(X_sdn)].reset_index(drop=True)
X_pub_scaled = X_scaled.iloc[len(X_sdn):].reset_index(drop=True)

y_sdn = y_sdn.reset_index(drop=True)
y_pub = y_pub.reset_index(drop=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

model = RandomForestClassifier(n_estimators=100)

model.fit(X_pub_scaled, y_pub)

y_pred = model.predict(X_sdn_scaled)

print(classification_report(y_sdn, y_pred))
```

```python
print("Actual:")
print(y_sdn.value_counts())

print("\nPredicted:")
print(pd.Series(y_pred).value_counts())
```
